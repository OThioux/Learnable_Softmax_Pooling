```python
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
from tqdm import tqdm
import cv2
```

```python
def create_model_with_spatial_scores(base_model, num_classes):
    """
    Modifies a given model to output spatial score maps for Spatial Consistency Loss.

    Args:
        base_model (tf.keras.Model): Backbone model without the top dense layers.
        num_classes (int): Number of output classes.

    Returns:
        tf.keras.Model: Modified model that outputs spatial score maps.
    """
    # Extract the feature maps from the base model
    feature_maps = base_model.output  # Shape: [batch_size, H, W, C]

    # Add a 1x1 convolutional layer to produce spatial score maps
    spatial_scores = tf.keras.layers.Conv2D(num_classes, (1, 1), activation='sigmoid', name='spatial_scores')(feature_maps)

    # Optionally add global average pooling for class-level predictions
    class_scores = tf.keras.layers.GlobalAveragePooling2D(name='class_scores')(spatial_scores)

    # Create the model
    model = tf.keras.models.Model(inputs=base_model.input, outputs=[spatial_scores, class_scores])
    return model
```

```python
# Example Usage
# Load a backbone model without the top dense layers
base_model = tf.keras.applications.ResNet50V2(include_top=False, weights='imagenet')

# Modify the model for 10 classes
num_classes = 10
model = create_model_with_spatial_scores(base_model, num_classes)

# Print model summary
model.summary()
```

```python
import tensorflow as tf

class TransformationManager:
    """
    A class to manage transformations for Spatial Consistency Loss.
    Handles recording, applying, and reversing transformations to ensure proper heatmap updates.
    """
    def __init__(self, crop_fraction=0.8, horizontal_flip=True):
        """
        Initialize the TransformationManager with specified parameters.

        Args:
            crop_fraction (float): Fraction of the image to retain during cropping (0 < crop_fraction <= 1).
            horizontal_flip (bool): Whether to apply horizontal flipping.
        """
        self.crop_fraction = crop_fraction
        self.horizontal_flip = horizontal_flip
        self.transformations = {}

    def apply_transformations(self, image, image_id):
        """
        Apply transformations to the image and record the operations.

        Args:
            image (tf.Tensor): Input image, shape [height, width, channels].
            image_id (hashable): Unique identifier for the image (e.g., filename or hash).

        Returns:
            tf.Tensor: Transformed image.
        """
        # Original dimensions
        original_height, original_width = tf.shape(image)[0], tf.shape(image)[1]

        # Random cropping
        crop_height = tf.cast(self.crop_fraction * tf.cast(original_height, tf.float32), tf.int32)
        crop_width = tf.cast(self.crop_fraction * tf.cast(original_width, tf.float32), tf.int32)
        offset_height = tf.random.uniform([], 0, original_height - crop_height + 1, dtype=tf.int32)
        offset_width = tf.random.uniform([], 0, original_width - crop_width + 1, dtype=tf.int32)
        cropped_image = tf.image.crop_to_bounding_box(image, offset_height, offset_width, crop_height, crop_width)

        # Random horizontal flipping
        flip = tf.random.uniform([]) > 0.5
        flipped_image = tf.cond(flip, lambda: tf.image.flip_left_right(cropped_image), lambda: cropped_image)

        # Record transformations
        self.transformations[image_id] = {
            "crop": (offset_height, offset_width, crop_height, crop_width),
            "flip": flip,
            "original_size": (original_height, original_width)
        }

        return flipped_image

    def reverse_transformations(self, heatmap, image_id):
        """
        Reverse the recorded transformations on the heatmap to align it with the original image space.

        Args:
            heatmap (tf.Tensor): Heatmap corresponding to the transformed image, shape [H, W, num_classes].
            image_id (hashable): Unique identifier for the image (e.g., filename or hash).

        Returns:
            tf.Tensor: Heatmap aligned with the original image space, updated only for visible regions.
        """
        if image_id not in self.transformations:
            raise ValueError(f"No recorded transformation found for image_id: {image_id}")

        transform = self.transformations[image_id]
        offset_height, offset_width, crop_height, crop_width = transform["crop"]
        flip = transform["flip"]
        original_height, original_width = transform["original_size"]

        # Resize heatmap to match cropped region in the original space
        heatmap_resized = tf.image.resize(heatmap, [crop_height, crop_width], method='bilinear')

        # Create a blank heatmap in the original size
        blank_heatmap = tf.zeros([original_height, original_width, tf.shape(heatmap)[-1]])

        # Place the resized heatmap back in the cropped location
        updated_heatmap = tf.tensor_scatter_nd_update(
            blank_heatmap,
            indices=tf.reshape(tf.range(offset_height, offset_height + crop_height)[:, None] * original_width +
                               tf.range(offset_width, offset_width + crop_width), [-1, 1]),
            updates=tf.reshape(heatmap_resized, [-1, tf.shape(heatmap_resized)[-1]])
        )

        # Reverse the flipping transformation
        if flip:
            updated_heatmap = tf.image.flip_left_right(updated_heatmap)

        return updated_heatmap
```
